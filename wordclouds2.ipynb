```python
import pandas as pd
import io
  
df = pd.read_csv(io.BytesIO(uploaded['df (4).csv']))
df = df.astype(str)
print(df)
```

```python
!pip install wordcloud
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
profs = []
for x in df.professions:
  profs.append(x)
print(profs)
#wordcloud = WordCloud().generate(" ".join(profs))
```

```python
vpdf = df[df.type.eq("Vice President")]
print(vpdf)
```

```python
vpwordcloud = vpdf.professions
vpwordcloud
```

```python
vp_profs = []
for x in vpwordcloud:
  vp_profs.append(x)
wordcloud = WordCloud(width = 600, height = 600, background_color = "white").generate(" ".join(vp_profs))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
wordcloud.to_file("vp_cloud.png")
print(wordcloud.words_.keys())
```

```python
pdf = df[df.type.eq("President")]
print(pdf)
```

```python
pwordcloud = pdf.professions
pwordcloud
```

```python
p_profs = []
for x in pwordcloud:
  p_profs.append(x)
wc = WordCloud(width = 600, height = 600, background_color = "white").generate(" ".join(p_profs))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
wc.to_file("presidents_cloud.png")
print(wc.words_.keys())
```
